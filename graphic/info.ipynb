```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/emilytsen/Documents/projects/ds_cancer/df/df_countrys.csv')
```

```python
# Rename columns to remove whitespace
data.columns = data.columns.str.strip().str.replace(' ', '_')
```

```python
year_min = data['YEAR'].min()
year_max = data['YEAR'].max()
print(f"The data go from year {year_min} to year {year_max}")
```

```python
# Filter data to include only relevant columns
stomach_cancer_data = data[['YEAR', 'STOMACH CANCER']].copy()

# Group by year and sum the number of stomach cancer deaths
stomach_cancer_yearly = stomach_cancer_data.groupby('YEAR')['STOMACH CANCER'].sum().reset_index()
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(stomach_cancer_yearly['YEAR'], stomach_cancer_yearly['STOMACH CANCER '], marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.title('Number of Deaths from Stomach Cancer Over the Years')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
