
### Filter rows where 'Code' is null.

```python
rows_without_code = data[data['Code'].isnull()]

rows_without_code
```

### Replace the code for "North America" with "NAM"

```python
data.loc[data['Country'] == 'North America', 'Code'] = 'NAM'
data.loc[data['Country'] == 'Caribbean', 'Code'] = 'CAR'
```

```python
rows_without_code = data[data['Code'].isnull()]

rows_without_code
```

```python
unique_countries = data['Country'].unique()

unique_countries
```

```python
# view the columns available in the DF
print(data.columns)

```

```python
# List of names you want to remove
remove_name = ['American Samoa', 'Andean Latin America', 'Australasia', 'Caribbean',
                   'Central Asia', 'Central Europe', 'Central Latin America', 
                   'Central Sub-Saharan Africa', 'East Asia', 'Eastern Europe', 
                   'Eastern Sub-Saharan Africa', 'Latin America and Caribbean', 
                   'Micronesia (country)', 'North America', 'North Africa and Middle East', 
                   'Northern Mariana Islands', 'Oceania', 'Palestine', 'South Asia',
                   'South Sudan', 'Southeast Asia', 'Southern Latin America', 
                   'Southern Sub-Saharan Africa', 'Sub-Saharan Africa', 
                   'Tropical Latin America', 'Western Europe', 
                   'Western Sub-Saharan Africa', 'World', 'United Kingdom']

# Remove rows corresponding to the names in the list
filtered_data = data[~data['Country'].isin(remove_name)]

# Now 'filtered_data' contains only the rows that are not in the list
```

```python
countries = filtered_data['Country'].unique()

countries
```

```python
filtered_data.columns = filtered_data.columns.str.upper()

filtered_data
```

```python
filtered_data['COUNTRY'] = filtered_data['COUNTRY'].str.upper()
print(filtered_data)
```

```python
# Round all numbers to 2 decimal places
filtered_data = filtered_data.round(2)
filtered_data
```

```python
filtered_data['YEAR'].dtype
```

```python
filtered_data['YEAR'] = pd.to_datetime(filtered_data['YEAR'], format='%Y').dt.year
```
